```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV files
yolov5_data = pd.read_csv('yolov5_data.csv')
yolov8_data = pd.read_csv('yolov8_data.csv')

# Extract relevant columns for YOLOv5
x5 = yolov5_data['Params (M)']
y5 = yolov5_data['mAPval 50-95']
z5_cpu = yolov5_data['Speed CPU b1 (ms)']
z5_gpu = yolov5_data['Speed V100 b1 (ms)']

# Extract relevant columns for YOLOv8
x8 = yolov8_data['Params (M)']
y8 = yolov8_data['mAPval 50-95']
z8_cpu = yolov8_data['Speed CPU ONNX (ms)']
z8_gpu = yolov8_data['Speed A100 TensorRT (ms)']

# Normalize the inference times for dot sizes
max_time = max(z5_cpu.max(), z5_gpu.max(), z8_cpu.max(), z8_gpu.max())
dot_size_5_cpu = (z5_cpu / max_time) * 100
dot_size_5_gpu = (z5_gpu / max_time) * 100
dot_size_8_cpu = (z8_cpu / max_time) * 100
dot_size_8_gpu = (z8_gpu / max_time) * 100

# Increase factor for dot sizes
factor_increase = 10

# Color scheme
colors = {
    'yolov5_cpu': '#1f77b4',  # Cooler blue
    'yolov5_gpu': '#2ca02c',  # Green
    'yolov8_cpu': '#d62728',  # Cooler red
    'yolov8_gpu': '#9467bd'   # Purple
}

# Create the plots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Params vs mAP Plot with connected dots
axs[0].scatter(x5, y5, alpha=0.6, color=colors['yolov5_cpu'], label='YOLOv5', edgecolor='black', marker='o')
axs[0].scatter(x8, y8, alpha=0.6, color=colors['yolov8_cpu'], label='YOLOv8', edgecolor='black', marker='o')
axs[0].plot(x5, y5, color=colors['yolov5_cpu'], alpha=0.7)
axs[0].plot(x8, y8, color=colors['yolov8_cpu'], alpha=0.7)
axs[0].set_xlabel('Params (M)')
axs[0].set_ylabel('mAPval 50-95')
axs[0].set_title('Params vs mAP')
axs[0].legend()
axs[0].grid(True)

# Params vs Time Plot with connected dots
axs[1].scatter(x5, dot_size_5_cpu * factor_increase, alpha=0.6, color=colors['yolov5_cpu'], label='YOLOv5 CPU Time', edgecolor='black', marker='o')
axs[1].scatter(x5, dot_size_5_gpu * factor_increase, alpha=0.6, color=colors['yolov5_gpu'], label='YOLOv5 GPU Time', edgecolor='black', marker='o')
axs[1].scatter(x8, dot_size_8_cpu * factor_increase, alpha=0.6, color=colors['yolov8_cpu'], label='YOLOv8 CPU Time', edgecolor='black', marker='o')
axs[1].scatter(x8, dot_size_8_gpu * factor_increase, alpha=0.6, color=colors['yolov8_gpu'], label='YOLOv8 GPU Time', edgecolor='black', marker='o')
axs[1].plot(x5, dot_size_5_cpu * factor_increase, color=colors['yolov5_cpu'], alpha=0.7)
axs[1].plot(x5, dot_size_5_gpu * factor_increase, color=colors['yolov5_gpu'], alpha=0.7)
axs[1].plot(x8, dot_size_8_cpu * factor_increase, color=colors['yolov8_cpu'], alpha=0.7)
axs[1].plot(x8, dot_size_8_gpu * factor_increase, color=colors['yolov8_gpu'], alpha=0.7)
axs[1].set_xlabel('Params (M)')
axs[1].set_ylabel('Inference Time (Scaled)')
axs[1].set_title('Params vs Time')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()

```
